```
def trapezioSR(f,a,b):
    ''' Aplica o método do trapézio simples para aproximar a integral de f no intervalo [a,b]'''
    return (b-a)*(f(a)+f(b))/2
```

# Método 3/8 de Simpson (sem repetição)

```
def simpson3_8SR(f,a,b):
  ''' Aplica o método de 1/3 de Simpson para aproximar a integral de f no intervalo [a,b]'''
  m1 = (2 * a + b) / 3
  m2 = (a + 2 * b) / 3
  return (b - a) / 8 * (f(a) + 3 * f(m1) + 3 * f(m2) + f(b))
```

```
def f(x):
  return (x**5)-2*(x**2)+1

simpson3_8SR(f, 0, 2)
```

# Método dos Trapézios Repetidos




```
import numpy as np

def metodoTrapezioR(a, b, fx, n):
  h = (b - a)/n
  x = np.zeros(n + 1)
  y = np.zeros(n + 1)
  new_y = np.zeros(n + 1)
  # Construindo o vetor de intervalos para realizar o somatório
  x[0] = a
  for i in range(a, n + 1):
    x[i] = a + (i*h)
  
  # Utilizando os valores encontrados em x para trocar na função associada
  for i in range(a, n + 1):
    y[i] = fx(x[i])
  
  """
  Realizando a multiplicação por 2 dos valores do somatório, pois ao realizar o f(x)+f(x+1)
  os valores entre esses intervalos vão ser somados duas vezes 

  -- como exemplo 0 a 1 => utilizando um subintervalo de 10 valores de 0.1 quando for feito f(x) + f(x+1)
  ele se comportará da seguinte forma:

    f(0) + f(0.1)
    f(0.1) + f(0.2)
    f(0.2) + f(0.3)
    f(0.3) + f(0.4)
    .
    .
    .
  """
  for i in range(a, n + 1):
    if (i == 0):
      new_y[i] = y[i]
    
    elif (i == n):
      new_y[i] = y[n]
    
    else:
      new_y[i] = 2*y[i]
  
  somatorio = (h/2) * sum(new_y)
  
  return somatorio
```

```
def f(x):
  return (x**5)-2*(x**2)+1

metodoTrapezioR(0, 2, f, 10)
```

# Método 1/3 de Simpson Repetido

```
def simpson1_3(a, b, fx, n):
  h = (b - a)/n
  x = np.zeros(n + 1)
  y = np.zeros(n + 1)
  new_y = np.zeros(n + 1)
  # Construindo o vetor de intervalos para realizar o somatório
  x[0] = a
  for i in range(a, n + 1):
    x[i] = a + (i*h)

  for i in range(a, n + 1):
    y[i] = fx(x[i])
  

  for i in range(a, n + 1):
    if (i == 0):
      new_y[i] = y[i]
    
    elif (i == n):
      new_y[i] = y[n]
    
    elif ((i%2) == 0):
      new_y[i] = 2*y[i]
    
    else:
      new_y[i] = 4*y[i]

  somatorio = (h/3) * sum(new_y)
  
  return somatorio
```

```
def f(x):
  return (x**5)-2*(x**2)+1

simpson1_3(0, 2, f, 10)
```

# Método 3/8 de Simpson Repetido

```
def simpson3_8(a, b, fx, n):
  h = (b - a)/n
  x = np.zeros(n + 1)
  y = np.zeros(n + 1)
  new_y = np.zeros(n + 1)

  x[0] = a
  for i in range(0, n + 1):
    x[i] = a + (i*h)
  
  for i in range(0, n + 1):
    y[i] = fx(x[i])
  
  for i in range(0, n + 1):
    if(i == 0):
      new_y[i] = y[i]
    
    elif(i == n):
      new_y[i] = y[n]

    elif(i == 1 or i == 2):
      new_y[i] = 3*y[i]

    elif((i%3) == 0):
      new_y[i] = 3*y[i]

    else:
      new_y[i] = 2*y[i]


  somatorio = (3/8 * h) * (sum(new_y))
  
  return somatorio
```

```
def f(x):
  return (x**5)-2*(x**2)+1

simpson3_8(0, 2, f, 10)
```

#  Aproximação por Diferenças Centrada

```
def f(x):
  return pow(x,3) + pow(x,2) + x + 3
def derivada(x,h):
  deriv = (f(x+h) - f(x-h))/(2*h)
  print("Primeira derivada com O(h²)", deriv)
def derivadao4(x,h):
  deriv = (-f(x+h*2)+8*f(x+h)-8*f(x-h)+f(x-h*2))/(12*h)
  print("Primeira derivada com O(h4)", deriv)
def segundaDerivada(x,h):
  deriv = (f(x+h) -2*f(x)+f(x-h))/h**2
  print("Segunda derivada com O(h²)", deriv)
def segundaDerivadao4(x,h):
  deriv = (-f(x+h*2)+16*f(x+h)-30*f(x)+16*f(x-h)-f(x-h*2))/12*h**2
  print("Segunda derivada com O(h4)", deriv)
def terceiraDerivada(x,h):
  deriv = (f(x+h*2)-2*f(x+h)+2*f(x-h)-f(x-h*2))/2*h**3
  print("Terceira derivada com O(h²)", deriv)
def terceiraDerivadao4(x,h):
  deriv = (-f(x+h*3)+8*f(x+h*2)-13*f(x+h)+13*f(x-h)-8*f(x-h*2)+f(x-h*3))/8*h**3
  print("Terceira derivada com O(h4)", deriv)
def quartaDerivada(x,h):
  deriv = (f(x+h*2)-4*f(x+h)+6*f(x)-4*f(x-h)+f(x+h*2))/h**4
  print("Quarta derivada com O(h²)", deriv)
def quartaDerivadao4(x,h):
  deriv = (-f(x+h*3)+12*f(x+h*2)+39*f(x+h)+56*f(x)-39*f(x-h)+12*f(x-h*2)+f(x-h*3))/6*h**4
  print("Quarta derivada com O(h4)", deriv)
  
derivada(3,0.25)
derivadao4(3, 0.25)
segundaDerivada(3,0.25)
segundaDerivadao4(3, 0.25)
terceiraDerivada(3,0.25)
terceiraDerivadao4(3, 0.25)
quartaDerivada(3,0.25)
quartaDerivadao4(3, 0.25)
```

```
import math
valor_real = 16.56637

def f(x):
  return 8 + 4*math.cos(x)

integral = trapezioSR(f, 0, math.pi/2)
erro = np.abs(valor_real - integral)/valor_real * 100
print("A integral encontrada pelo método do trapézio sem repetição foi %f, seu erro relativo percentual é %f%%"  %(integral, erro))
```

Trata-se do caso mais simples de Fórmula de Newton-Cotes fechada, a  ideia  da  regra  do  trapézio  é  aproximar  a  função  f(x)  por  um  polinômio  de  ordem  1  (reta), nessa aproximação a integral da função f(x) pode ser aproximada pela área de 1 trapézio. Neste caso, consideramos n=1 e temos dois nós de integração

Utilizando o método do trapézio implementado acima, através de uma única aplicação, obtemos um erro relativo percentual de aproximadamente 5%.

```
import math
valor_real = 16.56637

def f(x):
  return 8 + 4*math.cos(x)

integral = metodoTrapezioR(0, math.pi/2, f, 2)
erro = np.abs(valor_real - integral)/valor_real * 100
print("A integral encontrada pelo método do trapézio com n = 2 foi %f, seu erro relativo percentual é %f%%"  %(integral, erro))

integral = metodoTrapezioR(0, math.pi/2, f, 4)
erro = np.abs(valor_real - integral)/valor_real * 100
print("A integral encontrada pelo método do trapézio com n = 4 foi %f, seu erro relativo percentual é %f%%"  %(integral, erro))
```

Como é claro, se usassemos a regra dos trapézios simples para calcular o integral num intervalo [a, b], iamos obter uma aproximação grosseira do valor do integral, por isso, interessa-nos decompor esse intervalo em sub-intervalos cada vez mais pequenos, e nesses sub-intervalos aplicamos a regra dos trapézios simples.
Trata-se, neste caso, de fazer uma aproximação, da função integranda, usando um spline linear. 

Para simplificar, consideramos que o tamanho desses sub-intervalos é constante = h.
Assim, definimos h = ( b - a ) / N, onde N é o número de sub-intervalos ( = número de nós - 1), portanto, o valor do integral é igual à soma dos integrais nos sub-intervalos.

```
import math
valor_real = 16.56637

def f(x):
  return 8 + 4*math.cos(x)

integral = simpson1_3SR(f, 0, math.pi/2)
erro = np.abs(valor_real - integral)/valor_real * 100
print("A integral encontrada pelo método de 1/3 de simpson sem repetição foi %f, seu erro relativo percentual é %f%%"  %(integral, erro))
```

Tal como a Regra dos Trapézios, trata-se de outro exemplo de Fórmula de Newton-Cotes fechada, mas, ao invés de considerarmos a aproximação em cada sub-intervalo através de um polinómio interpolador do 1º grau (recta), podemos pensar numa aproximação um pouco melhor, considerando um polinómio interpolador do 2º grau (parábola).

A regra 1/3 de Simpson exatamente por utilizar parábola para aproximar a função no intervalo de integração tem como vantagem um erro menor do que a integral calculada pela regra do trapézio.

Para calcular o polinômio de segundo grau é necessário ter 3 pontos que interpolem a função. Dois pontos são obtidos nas extremidades.

O ponto x1 pode ser obtido de qualquer valor x no intervalo [a,b]. Por convenção, faz-se x1​ o ponto médio do intervalo. 

```
import math
valor_real = 16.56637
def f(x):
  return 8 + 4*math.cos(x)

integral = simpson1_3(0, math.pi/2, f, 4)
erro = np.abs(valor_real - integral)/valor_real * 100
print("A integral encontrada pelo método de 1/3 de simpson com n = 4 foi %f, seu erro relativo percentual é %f%%"  %(integral, erro))
```

A Regra 1/3 de Simpson repetida consiste em obtermos um intervalo com n pares de subintervalos, de modo que tenhamos m = 2n subintervalos. O tamanho de cada subintervalo é definido por h = (b - a)/m.

Comparando os resultados do método repetido e o não repetido, vemos que o erro se mostra bem menor onde temos a repetição. Se quisermos ter uma aproximação ainda melhor, basta aumentarmos o valor de n o quanto desejarmos. 

```
import math
import numpy as np
valor_real = 16.56637

def f(x):
  return 8 + 4*math.cos(x)

integral = simpson3_8SR(f, 0, math.pi/2)
erro = np.abs(valor_real - integral)/valor_real * 100
print("A integral encontrada pelo método de 3/8 sem repetição foi %f, seu erro relativo percentual é %f%%"  %(integral, erro))
```

Na regra regra 1/3 de Simpson a função f(x) é aproximada por um polinômio de segundo grau no intervalo de integração [a,b]. Já na regra 3/8 de Simpson, aproxima-se a função por um polinômio de terceiro grau, de forma que são necessários 4 pontos para obter esse polinômio por interpolação. Para ter os pontos, divide-se o intervalo [a,b] em três subintervalos igualmente espaçados de h

```
import math
def f(x):
  return 8 + 4*math.cos(x)

integral = simpson3_8(0, math.pi/2, f, 5)
erro = np.abs(valor_real - integral)/valor_real * 100
print("A integral encontrada pelo método de 3/8 com n = 5 foi %f, seu erro relativo percentual é %f%%"  %(integral, erro))
```

Use aproximações por diferenças centradas para obter estimativas para a primeira e a segunda derivadas de y = exem x = 2 para h = 0,1. Empregue tanto fórmulas de ordem de O(h2) como de ordem O(h4) para suas estimativas. 

```
import math

def f(x):
  return math.e**x
  
def derivada(x,h):
  deriv = (f(x+h) - f(x-h))/(2*h)
  print("Primeira derivada com O(h²)", deriv)
def derivadao4(x,h):
  deriv = (-f(x+h*2)+8*f(x+h)-8*f(x-h)+f(x-h*2))/(12*h)
  print("Primeira derivada com O(h4)", deriv)
def segundaDerivada(x,h):
  deriv = (f(x+h) -2*f(x)+f(x-h))/h**2
  print("Segunda derivada com O(h²)", deriv)
def segundaDerivadao4(x,h):
  deriv = (-f(x+h*2)+16*f(x+h)-30*f(x)+16*f(x-h)-f(x-h*2))/12*h**2
  print("Segunda derivada com O(h4)", deriv)
def terceiraDerivada(x,h):
  deriv = (f(x+h*2)-2*f(x+h)+2*f(x-h)-f(x-h*2))/2*h**3
  print("Terceira derivada com O(h²)", deriv)
def terceiraDerivadao4(x,h):
  deriv = (-f(x+h*3)+8*f(x+h*2)-13*f(x+h)+13*f(x-h)-8*f(x-h*2)+f(x-h*3))/8*h**3
  print("Terceira derivada com O(h4)", deriv)
def quartaDerivada(x,h):
  deriv = (f(x+h*2)-4*f(x+h)+6*f(x)-4*f(x-h)+f(x+h*2))/h**4
  print("Quarta derivada com O(h²)", deriv)
def quartaDerivadao4(x,h):
  deriv = (-f(x+h*3)+12*f(x+h*2)+39*f(x+h)+56*f(x)-39*f(x-h)+12*f(x-h*2)+f(x-h*3))/6*h**4
  print("Quarta derivada com O(h4)", deriv)
  
derivada(2,0.1)
derivadao4(2, 0.1)
segundaDerivada(2,0.1)
segundaDerivadao4(2, 0.1)
```

O método da aproximação por diferenças centradas é uma das fórmulas que se baseia na aproximação de derivadas por diferenças finitas. A fórmula de aproximação obtém-se da série de Taylor da função derivada.

A partir das aproximações por diferença-quociente para derivadas de qualquer ordem, é possível transformar equações diferenciais em problemas lineares. Para isso, é necessário ignorar o termo de erro e tornar h um número muito pequeno, mas grande o suficiente para que não cause instabilidades nas aproximações das derivadas.
